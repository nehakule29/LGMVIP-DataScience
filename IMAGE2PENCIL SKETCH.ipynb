Task 4 - Image to Pencil Sketch

```python
from google.colab.patches import cv2_imshow
import cv2
import matplotlib.pyplot as plt
from skimage import io
```

```python
img=io.imread('/shahid_kapoor1[1].jpg',)
cv2_imshow(img)
```

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
cv2_imshow(gray_img)
```

```python
inverted_img = 255 - gray_img
cv2_imshow(inverted_img)
```

```python
blurred = cv2.GaussianBlur(inverted_img, (21,21), 0)
```

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_img, inverted_blurred, scale=256.0)
```

```python
cv2_imshow(pencil_sketch)
```
